```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['LotArea']),
    ('tnf2',OrdinalEncoder(categories=[['1Fam', 'TwnhsE','Duplex','Twnhs','2fmCon']]),['BldgType']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['LotConfig', 'Exterior1st'])
],remainder='passthrough')
```
