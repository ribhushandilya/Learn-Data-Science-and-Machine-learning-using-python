```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ['area_type', 'availability', 'society', 'balcony'], axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['location'].value_counts()
```

```python
df['location'] = df['location'].fillna('Sarjapur Road')
# df['location'].dropna(axis = 0, inplace  =True)
```

```python
df.isnull().sum()
```

```python
df['size'].value_counts()
```

```python
df['size'] = df['size'].fillna('2 BHK')
```

```python
df['bath'] = df['bath'].fillna(df['bath'].median())
```

```python
df.isnull().sum()
```

```python
df['bhk'] = df['size'].str.split().str.get(0).astype(int)
df.drop(columns = 'size', axis = 1, inplace = True)
```

```python
df.head()
```

```python
df[df.bhk > 20]
```

```python
df['total_sqft'].unique()
```

```python
def change(x):
    temp = x.split('-')
    if len(temp)==2:
        return(float(temp[0]))+(float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(change)
```

```python
df.describe()
```

```python
df['ppsft'] = df['price']*100000 / df['total_sqft']
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
### Use Lambda function to remove extra spaces
df['location']=df['location'].apply(lambda x:x.strip())
location_count=df['location'].value_counts()
```

```python
df['location'].value_counts()
```

```python
location_count_less_10=location_count[location_count<=10]
location_count_less_10
```

```python
df['location'] = df['location'].apply(lambda x: 'other' if x in location_count_less_10 else x)
```

```python
df['location'].value_counts()
```

```python
df.describe()
```

```python
df = df[((df['total_sqft']/df['bhk'])>300)]
df.describe()
```

```python
upper_limit=df['ppsft'].quantile(0.99)
lower_limit=df['ppsft'].quantile(0.01)
print(upper_limit)
print(lower_limit)
```

```python
df = df[((df['ppsft']<=20789.81) & (df['ppsft']>=2500.0))]
df
```

```python
df.drop(columns = 'ppsft', inplace = True)
```

```python
df
```

```python
df.to_csv('Cleaned_bengaluru_House_prediction.csv')
```

```python
X = df.drop(columns = ['price'])
X.head()
```

```python
y = df[['price']]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
scaler = StandardScaler()
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LinearRegression, Lasso, Ridge
lr = LinearRegression()

column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['location']), 
    remainder='passthrough')

```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
y_pred_lr
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
print('r2_score',r2_score(y_test, y_pred_lr),'\n','mean_absolute_error',mean_absolute_error(y_test, y_pred_lr))
```

```python
lasso = Lasso()
ridge = Ridge()
```

```python
pipe=make_pipeline(column_trans,scaler,lasso) 
pipe.fit(X_train, y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lasso)
```

```python
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
```

```python
pipe = make_pipeline(column_trans,scaler,dtr)
pipe.fit(X_train, y_train)
```

```python
y_pred_dtr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_dtr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 10)

pipe = make_pipeline(column_trans,scaler,rf)
pipe.fit(X_train, y_train)
```

```python
y_pred_rf=pipe.predict(X_test)
r2_score(y_test,y_pred_rf)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors = 3)
```

```python
pipe = make_pipeline(column_trans,scaler,knn)
pipe.fit(X_train, y_train)
```

```python
y_pred_knn = pipe.predict(X_test)
r2_score(y_test,y_pred_knn)
```
