# Using normal Regression

```python
# Bengaluru_House_Data.csv
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(labels = ['society'], axis = 1, inplace = True)
```

```python
df['balcony'] = df['balcony'].bfill()
```

```python
df['bath'] = df['bath'].fillna(df['bath'].mean())
```

```python
df['location'] = df['location'].fillna(df['location'].mode()[0])
```

```python
df['size'] = df['size'].fillna(df['size'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df[['bath','balcony', 'price',]] = df[['bath','balcony', 'price']].astype(int)
df
```

```python
# df['location'].unique()
df.drop(labels = (['availability']),axis = 1, inplace = True)
```

```python
df['area_type'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['area_type'] = le.fit_transform(df['area_type'])
df['size'] = le.fit_transform(df['size'])
df['location'] = le.fit_transform(df['location'])

```

```python
import re

def convert_range(value):
    cleaned_value = re.sub(r'[^\d\s-]', '', value).strip()  
    if ' - ' in cleaned_value: 
        numbers = cleaned_value.split(' - ')
        min_value = float(numbers[0])
        max_value = float(numbers[1])
        return (min_value + max_value) / 2
    else:
        return float(cleaned_value)

df['total_sqft'] = df['total_sqft'].apply(convert_range)
```

```python
X = df.iloc[:,:-1]
X
```

```python
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.2)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scale  = StandardScaler()
```

```python
scale.fit(X_train)
```

```python
x_trained_scale = scale.transform(X_train)
x_test_scale = scale.transform(X_test)
```

```python
scale.mean_
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_trained_scale,y_train)
```

```python
y_pred = lr.predict(x_test_scale)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
r2_score_ = r2_score(y_test, y_pred)
print(r2_score_)
```

```python
mae = mean_absolute_error(y_test, y_pred)
print(mae)
```

# Using Pipeline and scalling and column transfrmation

```python
import pandas as pd
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(columns=['area_type','availability','society','balcony'])
```

```python
df.describe()
```

```python
df['location']=df['location'].fillna('Sarjapur  Road')
```

```python
df['size'].value_counts()
```

```python
df['size']=df['size'].fillna('2 BHK')
```

```python
df['bath']=df['bath'].fillna(df['bath'].median())
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['bhk']=df['size'].str.split().str.get(0).astype('int')
df[df.bhk > 20]
```

```python
df['total_sqft'].unique()
```

```python
def convert(x):
    temp=x.split('-')
    if len(temp)==2:
        return(float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None

df['total_sqft']=df['total_sqft'].apply(convert)
df.head()
```

```python
df['price_per_sqft']=df['price']*100000/df['total_sqft']
df['price_per_sqft']
```

```python
df.describe()
```

```python
df['location'].value_counts()
```

```python
df['location']=df['location'].apply(lambda x:x.strip())
location_count=df['location'].value_counts()
location_count
```

```python
location_count_less_10=location_count[location_count<=10]
location_count_less_10
```

```python
df['location']=df['location'].apply(lambda x:'other' if x in location_count_less_10 else x)
df['location'].value_counts()
```

## Handle Outliers

```python
new_df=df[(df['price_per_sqft']<=21707.94) & (df['price_per_sqft']>=2500.36)]
new_df
```

```python
new_df = new_df.drop(columns=['area_type','availability','society','balcony'])
```

```python
new_df.shape
```

```python
new_df = new_df.drop(columns=['size','price_per_sqft'])
```

```python
new_df
```

```python
X = new_df.drop(columns=['price'])
y = new_df[['price']]
```

```python
print(X.head())
print(('*')*50)
print(('*')*50)
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import make_column_transformer
scaler=StandardScaler()
```

```python
column_trans=make_column_transformer((OneHotEncoder(sparse_output=False),['location']),
                                    remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.pipeline import make_pipeline
lr=LinearRegression()
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr=pipe.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
print(r2_score(y_test,y_pred_lr))
print(mean_absolute_error(y_test,y_pred_lr))
```
