# Concatination of array

```python
ar1 = np.random.randint(1,10,9).reshape(3,3)
ar2 = np.random.randint(1,10,9).reshape(3,3)
```

```python
ar1
```

```python
ar2
```

```python
ar1+ar2
```

```python
np.concatenate((ar1,ar2))
```

```python
#concatinating vertically
np.concatenate((ar1,ar2),axis=1)
```

```python
#Concatenating horizontally 
np.concatenate((ar1,ar2),axis=0)
```

```python
#Vertical stack
np.vstack((ar1,ar2))
```

```python
#Horizontal Stack
np.hstack((ar1,ar2))
```

```python
np.split(ar1,3)
```

```python
x_sin = np.arange(1,5,0.20)
x_sin
```

```python
y_sin = np.sin(x_sin)
y_sin
```

```python
import matplotlib.pyplot as plt 
plt.scatter(x_sin,y_sin)
plt.xlabel("x_sin vales")
plt.ylabel("y_sin values")
plt.title("Sin Graph for x and y sin values")
plt.show()
```
