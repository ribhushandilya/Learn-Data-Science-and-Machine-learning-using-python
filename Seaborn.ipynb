#### In Seaborn, estimators are used to compute and represent aggregate values of data, especially when dealing with grouped or categorical data. Estimators are essentially statistical functions that help in summarizing and simplifying data to provide meaningful insights through visualizations.

```python
# estimator can be sum, None, mean, median, mode
plt.figure(figsize = (16,4))

sns.set(style = 'whitegrid')

sns.lineplot(x = 'tip', y = 'size',  data = tips_df,  hue = 'day', palette = 'deep', markers = 'day', ms = 3000, estimator = 'median')
plt.xlabel('Tip', fontsize = 15)
plt.ylabel('Size', fontsize = 15)
plt.title('Tips by size', fontsize = 20)
plt.show()
```

## Distplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.distplot(df['tip'])
# sns.histplot(df['tip'])
plt.show()
```

```python
# kde is used to show the pattern line

sns.distplot(df['total_bill'], kde = False)   
plt.show()
```

```python
sns.distplot(df['total_bill'], bins = 100)   
plt.show()
```

```python
sns.distplot(df['total_bill'], hist = False, bins = 100)   
plt.show()
```

```python
sns.distplot(df['total_bill'], rug = True, bins = 100)   
plt.show()
```

```python
from scipy.stats import norm
sns.distplot(df['total_bill'],fit=norm)
plt.show()
```

```python
sns.distplot(df['total_bill'],fit=norm, kde = False)
plt.show()
```

```python
sns.distplot(df['tip'], hist_kws = {'color': 'y','edgecolor':'k', 'linewidth':1}, fit = norm, fit_kws = {'color':'g', 'linestyle':'-', 'linewidth':4},kde_kws = {'color': 'r', 'linewidth' :1, 'linestyle':'--'})
plt.show()
```

```python
sns.distplot(df['tip'], hist_kws = {'color': 'y','edgecolor':'k', 'linewidth':1}, fit = norm,
             fit_kws = {'color':'g', 'linestyle':'-', 'linewidth':4}
             ,kde_kws = {'color': 'r', 'linewidth' :1, 'linestyle':'--'}, rug = True)
plt.show()
```

```python
from scipy.stats import norm
sns.distplot(df['total_bill'],fit=norm, axlabel = 'Total Bill', label = 'Total Bill')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15,8))

sns.distplot(df['tip'], label = 'Tip', bins = 100)
sns.distplot(df['total_bill'], label = 'Total Bill Paid')
sns.distplot(df['size'], label = 'Size')
plt.legend()
plt.show()
```
